```python
# =========================
# JAFFE Emotion Recognition - Single Script
# =========================

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Set dataset path (CHANGE THIS)
dataset_path = "/content/drive/MyDrive/jaffe (Unzipped files)"

# =========================
# Import Libraries
# =========================
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# =========================
# Load Dataset
# =========================
images = []
labels = []

emotion_map = {
    'AN': 0,  # Angry
    'DI': 1,  # Disgust
    'FE': 2,  # Fear
    'HA': 3,  # Happy
    'NE': 4,  # Neutral
    'SA': 5,  # Sad
    'SU': 6   # Surprise
}

for file in os.listdir(dataset_path):
    if file.endswith(".tiff"):
        img_path = os.path.join(dataset_path, file)

        # Read and preprocess image
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (48, 48))

        images.append(img)

        # Extract label (e.g., KA.HA1.39.tiff → HA)
        label_code = file.split('.')[1][:2]
        labels.append(emotion_map[label_code])

# Convert to numpy
X = np.array(images) / 255.0
X = X.reshape(-1, 48, 48, 1)
y = np.array(labels)

print("Dataset loaded:", X.shape)

# =========================
# Build Model
# =========================
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(48,48,1)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dense(7, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# =========================
# Train Model
# =========================
model.fit(X, y, epochs=15, batch_size=16, validation_split=0.2)

# =========================
# Test Prediction
# =========================
emotion_labels = ['Angry','Disgust','Fear','Happy','Neutral','Sad','Surprise']

index = 10  # change index to test different images
prediction = model.predict(X[index].reshape(1,48,48,1))

print("Predicted Emotion:", emotion_labels[np.argmax(prediction)])

# Show image
plt.imshow(X[index].reshape(48,48), cmap='gray')
plt.title("Prediction: " + emotion_labels[np.argmax(prediction)])
plt.axis('off')
plt.show()
```
